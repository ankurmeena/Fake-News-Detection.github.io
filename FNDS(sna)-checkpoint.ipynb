```python
import pandas as pd
import seaborn as sa
import matplotlib.pyplot as plt
```

```python
data_import = pd.read_csv('~/Desktop/FNDS/fsna.csv',index_col=0)
data_import.head()
```

```python
data_import.shape
data = data_import.drop(["title", "subject","date"], axis = 1)
data.isnull().sum()
```

```python
# Shuffling of data
data = data.sample(frac=1)
data.reset_index(inplace=True)
data.drop(["index"], axis=1, inplace=True)
```

```python
#plotting unique value based on each category fake and true
sa.countplot(data=data,
              x='class',
              order=data['class'].value_counts().index,palette = "Set1")
```

```python
from tqdm import tqdm
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud
```

```python
def function(text_data):
    preprocessed_text = []
      
    for sentence in tqdm(text_data):
        sentence = re.sub(r'[^\w\s]','', sentence)
        preprocessed_text.append(' '.join(token.lower()
                                  for token in str(sentence).split()
                                  if token not in stopwords.words('english')))
  
    return preprocessed_text
```

```python
preprocessed_data = function(data['text'].values)
data['text'] = preprocessed_data
```

```python
# combination of token that are said to be real
integrated = ' '.join(
    word for word in data['text'][data['class'] == 1].astype(str))
wordCloud = WordCloud(width=1600,
                      height=800,
                      random_state=21,
                      max_font_size=110,
                      collocations=False)
plt.figure(figsize=(15, 10))
plt.imshow(wordCloud.generate(integrated), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# combination of token that are said to be fake
integrated = ' '.join(
    word for word in data['text'][data['class'] == 0].astype(str))
wordCloud = WordCloud(width=1600,
                      height=800,
                      random_state=21,
                      max_font_size=110,
                      collocations=False)
plt.figure(figsize=(15, 10))
plt.imshow(wordCloud.generate(integrated), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#ploting the bargraph of the top 30 most frequent words
from sklearn.feature_extraction.text import CountVectorizer
  
  
def get_top_n_words(data, n=None):
    vec = CountVectorizer().fit(data)
    bag_of_words = vec.transform(data)
    sum_words = bag_of_words.sum(axis=0)
    words_freq = [(word, sum_words[0, idx])
                  for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1],
                        reverse=True)
    return words_freq[:n]
  
  
common_words = get_top_n_words(data['text'], 20)
df1 = pd.DataFrame(common_words, columns=['Review', 'count'])
  
df1.groupby('Review').sum()['count'].sort_values(ascending=False).plot(
    kind='bar',
    figsize=(10, 6),
    xlabel="Top Words",
    ylabel="Count",
    title="Bar Chart of Top 20 Words Frequency"
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
  
x_train, x_test, y_train, y_test = train_test_split(data['text'], 
                                                    data['class'], 
                                                    test_size=0.48)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
  
vectorization = TfidfVectorizer()
x_train = vectorization.fit_transform(x_train)
x_test = vectorization.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model1 = KNeighborsClassifier(n_neighbors = 1)
model1.fit(x_train, y_train)

# testing the model
print(accuracy_score(y_train, model1.predict(x_train)))
print(accuracy_score(y_test, model1.predict(x_test)))
```

```python
# Confusion matrix of Results from KNeighborsClassifier classification
from sklearn import metrics
from sklearn.metrics import classification_report
cm = metrics.confusion_matrix(y_test, model1.predict(x_test))
  
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,
                                            display_labels=[False, True])
  
cm_display.plot()
plt.show()
print(classification_report(y_test, model1.predict(x_test)))
```

AUC measures how well a model is able to distinguish between classes

```python
import numpy as np
from sklearn .metrics import roc_auc_score
 
auc = np.round(roc_auc_score(y_test,model1.predict(x_test)), 3)
 
print("Auc for our sample data is {}". format(auc))
```

```python
# importing machine learning models for prediction
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
model2.fit(x_train, y_train)

# testing the model
print(accuracy_score(y_train, model2.predict(x_train)))
print(accuracy_score(y_test, model2.predict(x_test)))
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
cm = metrics.confusion_matrix(y_test, model2.predict(x_test))
  
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,
                                            display_labels=[False, True])
  
cm_display.plot()
plt.show()
print(classification_report(y_test, model2.predict(x_test)))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import log_loss
# importing voting classifier
from sklearn.ensemble import VotingClassifier
 
# Making the final model using voting classifier
final_model = VotingClassifier(
    estimators=[('KNN', model1), ('RFC', model2)], voting='hard')
 
# training all the model on the train dataset
final_model.fit(x_train, y_train)
 
# predicting the output on the test dataset
pred_final = final_model.predict(x_test)
```

```python
# Confusion matrix of Results from voting classifier
from sklearn import metrics
from sklearn.metrics import classification_report
cm = metrics.confusion_matrix(y_test, pred_final)
  
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix=cm,
                                            display_labels=[False, True])
  
cm_display.plot()
plt.show()
print(classification_report(y_test,pred_final))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# y_true is the true labels, y_score1 and y_score2 are the predicted scores for each model
fpr1, tpr1, thresholds1 = roc_curve(y_test, model1.predict(x_test))
fpr2, tpr2, thresholds2 = roc_curve(y_test, model2.predict(x_test))
fpr3, tpr3, thresholds3 = roc_curve(y_test, pred_final)

# Calculate AUC for each model
auc1 = auc(fpr1, tpr1)
auc2 = auc(fpr2, tpr2)
auc3 = auc(fpr3, tpr3)

# Plot the ROC curves for each model
plt.plot(fpr1, tpr1, color='blue', label='Model 1 (AUC = %0.2f)' % auc1)
plt.plot(fpr2, tpr2, color='green', label='Model 2 (AUC = %0.2f)' % auc2)
plt.plot(fpr3, tpr3, color='red', label='final model (AUC = %0.2f)' % auc3)

# Set the axis labels and legend
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend(loc='lower right')

# Show the plot
plt.show()
```
